```python
def pmk_apt(placemark):
    '''
    Receives placemark element 
    returns list of apt info:
    returns name, description, lat, lon, etc
    '''
    pname=placemark.find('.//kml:name',namespaces).text
    desc=placemark.find('.//kml:description',namespaces)
    if desc is None:
        desc=''
    elif placemark.find('.//kml:description',namespaces).text == '&nbsp;':
        desc=''
    else:
        desc=placemark.find('.//kml:description',namespaces).text
    coor=placemark.find('.//kml:Point/kml:coordinates',namespaces).text
    return pname, desc, coor.split(',')[0], coor.split(',')[1]
    
def placemark_parser(el):
    xplace='./kml:Placemark'  #only looks for folders one level down
    for nel in el.findall(xplace,namespaces):
        print(pmk_apt(nel))
        
def folder_parser(el,el_name):
    '''
    parse folders starting at el 
    returns a database? for now prints
    if a folder only no entry to database!
    '''
    xfolder='./kml:Folder'  #only looks for folders one level down
    for nel in el.findall(xfolder,namespaces):
        name=el_name+','+nel[0].text
        print(name)
        placemark_parser(nel)
        if nel.find(xfolder,namespaces):
            folder_parser(nel,name)
```

```python
'''
Position data format
The position format is used in all data files. This format allows usage of 32 bit integers while
allowing position resolution to less than 3 feet.
Position latitude and longitude are stored as signed 32 bit integers.
Degrees are multiples of the value 180000 and any degree fraction is a multiple of 1/180000.
Degrees North and East are positive while degrees South and West are negative.
Example: North 45 degrees, 59 minutes, 30 seconds is 8278500.
Example: North 0 degrees, 30 minutes, 0 seconds is 90000.
'''
def show_dms(value):
    div=180000
    dd=value/div
    d= int(dd)
    md= (dd-d) * 60
    m= int(md) #if negative lat/long show it only in degrees not in minutes and seconds as well
    sd= (md-m) * 60
    s = int(sd)
    return(d, abs(m), abs(s))

def show_deg(value):
    div=180000
    dd=value/div
    return(dd)

def show_date(val):
    import datetime
    return(datetime.datetime(2000,1,1,0,0,0)+datetime.timedelta(seconds=val))

def show_airport(airport):
    tipe=['Unspecified','AIRPORT', 'MAJOR AIRPORT','SEAPLANE BASE','AIRFIELD',
          'PRIVATE AIRFIELD','ULTRALIGHT FIELD','HELIPORT']
    #print(airport['type'],tipe[airport['type']])
    print(airport['type'])
    print(airport['len'])
    print((airport['name'][:airport['len']]))
    print(show_deg(airport['lat']))
    print(show_deg(airport['lon']))
    print((airport['pointer']))

def show_waypoint(waypoint):
    tipe=['Unspecified','AIRPORT', 'MAJOR AIRPORT','SEAPLANE BASE','AIRFIELD',
          'PRIVATE AIRFIELD','ULTRALIGHT FIELD','HELIPORT']
    #print(airport['type'],tipe[airport['type']])
    print('Type: ',waypoint['type'])
    print('S len:',waypoint['short_len'])
    print('Sname:',waypoint['short_name'][:waypoint['short_len']])
    print('L len:',waypoint['long_len'])
    print('Lname:',waypoint['long_name'][:waypoint['long_len']])
    print('Lat:  ',show_deg(waypoint['lat']),show_dms(waypoint['lat']))
    print('Long: ',show_deg(waypoint['lon']),show_dms(waypoint['lon']))
    print('Data: ', (waypoint['data']))    
    print('Buffer:', (waypoint['buffer']))    
    
def show_obstacle(obstacle):
    Kind=['UNKNOWN','ARCH','BALLOON','BRIDGE','BUILDING','BUILDING/MAST','CATENARY','COOLING TOWER','CRANE',
          'CRANE T','CTRL TWR','DAM','DOME','ELEVATOR','MONUMENT','PLANT', 'POLE','RIG','REFINERY','SIGN',
          'SPIRE','CHIMNEY','CHIMNEYS','TANK','T-L TWR','TOWER','TOWERS','TRAMWAY','WINDMILL','WIND TURBINE',
          'SLAG HEAP','MAST','WATER TOWER','CHURCH','FLARE','TOWER/MAST','DUMP','CABLE','OBSTACLE','INDUSTRY',
          'VEGETATION', 'LIGHTHOUSE','CONTAINER','PYLON']
    Lighting=['Unknown','High Intensity White Strobe Lighting','Medium Intensity White Strobe Lighting',
              'Red Lighting','Dual, Red with HIGH Intensity White Strobe',
              'Dual, Red with MEDIUM Intensity White Strobe','Flood Lights','No Lights',
              ' Other, Lighting not listed']
    print('Lat:      ',show_deg(obstacle['lat']),show_dms(obstacle['lat']))
    print('Long:     ',show_deg(obstacle['lon']),show_dms(obstacle['lon']))
    print('Kind:     ',Kind[obstacle['Kind']])
    print('Lighting: ',Lighting[obstacle['Lighting']])
```

```python
#fileContent hardcoded. SLOPPY!

def get_header(value):
    #value is first 500 bytes of file =>fileContent[0:496+4])
    #cycle description is text
    #what is needed at the end the last 4 bytes?

    keys=['FileID', 'Vendor', 'NaviDataDate', 'MagicNumber1','MagicNumber2', 'Serials', 'NumberOfSections', 
    'EncryptionIdentifier', 'Version', 'NumberOfAirports', 'AirportIndex', 'Airports', 'NumberOfAirspaces', 'Airspaces', 
    'NumberOfAirways', 'VictorAirways', 'NumberOfSids', 'SIDS', 'NumberOfStars', 'STARS', 'NumberOfObstacles',
          'Obstacles','NumberOfWaypoints','Waypoints','WaypointAllocationTable','StartDate','EndDate','CycleStr',
          'NumberOfHoldingPatterns','HoldingPatterns','end']
    #value=struct.unpack('8siiii384sBBH13i2i3i4s2i4s',fileContent[0:496+4])
    headers=struct.unpack('8siiii384sBBH13i2i3i4s2i4s',value)
    return(dict(zip(keys, headers)))

def get_airport(value):
    '''
    Kind: byte (See airport types further on)
    Identifier: 1 byte length of string (1..6), followed by six ASCII characters
    Pointer: 32 bit signed integer
    Latitude: 32 bit signed integer
    Longitude: 32 bit signed integer
    Each index entry is exactly 20 bytes in size.
    '''
    tipe=['Unspecified','AIRPORT', 'MAJOR AIRPORT','SEAPLANE BASE','AIRFIELD',
          'PRIVATE AIRFIELD','ULTRALIGHT FIELD','HELIPORT']
    keys=['type','len','name','pointer','lat','lon']
    airport=struct.unpack('BB6siii',value)
    return(dict(zip(keys, airport)))

def get_waypoint(value):
    keys=['lat','lon','data','type','short_len','short_name','long_len','long_name','buffer']
    waypoint=struct.unpack('iiiBB6sB27si',value)
    return(dict(zip(keys, waypoint)))


def obs_header(address):
    obs_keys=['Version','Left','Bottom','Width','Height','LargestBIN','Check']
    record='5h2H'
    record_len=14
    obs_values=struct.unpack(record, fileContent[address:address+record_len])
    obs_head=dict(zip(obs_keys, obs_values))   
    return(obs_head)

def get_obstacle_bin(bin_addr):
    '''Input is header?'''
    record='iH'
    record_len=6
    bin_keys=['Pointer','Count']
    bin_values=struct.unpack(record,fileContent[bin_addr:bin_addr+record_len])
    bin_index=dict(zip(bin_keys,bin_values))
    return(bin_index)

def get_obstacles(value):
    obs_val=()
    obs_keys=['lat','lon','AGL','AMSL','Kind','Lighting']
    record_len=16
    record='iiHHHH'
    obs_val=struct.unpack(record,fileContent[value:value+record_len])
    obs=dict(zip(obs_keys,obs_val))
    return(obs)
    

'''
work in progress

?get all blocks of airspaces (in airspaces pointer from header)
    ~get all airspaces as individual airspace in (lat long block)
        -get one airspace complete with polygon (adress )
            -get fixed header of airspace
            -get variable strings of airspace
            -get all polygons of airspace
                -get variable coords of airspace polygon(s)
'''

###above functions work with snippet of fileContents
### below functions are variable in size and works with starting address to fileContents (still hardcoded!!!)

def get_airspace(address):
    '''
    Type - airspace type according to table
    - bounding box
    ptr_next - address offset to next airspace in block, 0 if none
    ptr_pts - address offset to start of airspace POLYGON
    '''

    head_keys=['type','nwlat','nwlon','selat','selon','ptr_next','ptr_pts','f1','f2','u_alt','l_alt']
    string_keys=['icao','name','class','exception','contact_name','level','times','wx']
    poly_keys=['points','coords']
    ###HEAD
    #get 11 integers fixed 
    record='11i'
    record_len=11*4
    airspace_head=struct.unpack(record,fileContent[address:address+record_len])
    
    ###STRINGS
    #get 8 possible strings
    len_list=[]
    str_list=[]
    run_off=address+record_len
    for airspace_string in range(8):
        next_len=((struct.unpack('B',fileContent[run_off:run_off+1])[-1]))
        run_off+=1
        next_str=((struct.unpack(str(next_len)+'s',fileContent[run_off:run_off+next_len])))
        run_off+=next_len
        len_list.append(next_len)
        str_list.append(next_str)
        
    ###POLYGON
    #get no. of polygon points
    points=((struct.unpack('i',fileContent[run_off:run_off+4])[-1]))
    run_off+=4 #get past no. of points to first coordinate then iterate over all pairs
    coord_str=[]
    for no,k in enumerate(range(0,points*8,8)):
        #print(no, k)
        lat=show_deg(struct.unpack('i',fileContent[run_off+k:run_off+k+4])[-1])
        lon=show_deg(struct.unpack('i',fileContent[run_off+k+4:run_off+k+8])[-1])
        coord_str.append((lat,lon))
        #if show_deg(lat)==(200,0,0) and show_deg(lon)==(0,0,0):
        #    print('*** End of polygon ***')
    poly_tuple=points, coord_str
    ###struct.unpack('i',fileContent[run_off+k+4+4:run_off+k+8+4])
    airspace=dict(zip(head_keys+string_keys+poly_keys, airspace_head+tuple(str_list)+poly_tuple))
    
    return(airspace)

def get_blk_airspace(offset, address):            #AIRSPACES from BLOCK starting with address
    while address: #gets first address of block and cycles through all in that block
        air_space=get_airspace(offset+address)
        print(air_space['icao'])
        print(air_space['name'])
        address=air_space['ptr_next']  #gets pointer to next airpace in block, last ptr will be 0
        print(address)
        
def get_block(offset):            #AIRSPACES from BLOCK starting with address
    record='i'
    record_len=4
    i=0
    for lat in range (90,-90,-10):
        for lon in range(-180,180,10):
            i=i+1  #position zero is just the id of the table to follow
            address=struct.unpack('i',fileContent[offset+i*record_len:offset+i*record_len+record_len])[0]
            #print('Blk: ',i ,'Lat: ', lon, 'Lon: ', lat, address)
            if address != 0: 
                print('Blk: ',i ,'Lat: ', lat, 'Lon: ', lon, address)
                get_blk_airspace(offset,address)
                #input()
```
