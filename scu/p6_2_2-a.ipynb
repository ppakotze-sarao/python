```python
#Import of Python available libraries
import sculib

#mpi = sculib.scu()
#mpi.ip = '10.96.64.10'
sim = sculib.scu()
sim.ip = '10.96.254.85'
```

```python
def mpi_t622a(scu):

    print('Test 6.2.2a')

    #Revisions:
    #0.1 initial write - could do with some kat sensor retrieval
    #0.2 rewrite for sculib

    #Common setup
    config_name = 'HN_TILT_TEST'

    
    loop1 = 5
    azimuths = [-180, -120, -60, 0, 60, 120, 180]
    #az_start = -180
    az_speed = 3

    el_bottom = 20 #instead of 15 due to temp limitation
    el_top = 85
    el_speed  = 0.1
    
    wait5 = 5
    wait10 = 10
    wait100 = 100
    sampling = 200
    scu.debug = False
    #debug test only remove these
    #'''
    loop1 = 5
    wait5 = 1
    wait10 = 1
    wait100 = 1
    sampling = 5000
    scu.debug = False
    #'''

    #B. Tilt Sensor tests [316-000000-043]
    '''
    (Note that this test is added to check the impact of elevation unbalance on the tilt)
    1) Manually move the Dish to 85° Elevation, -180° Azimuth and de-activate all axes
    2) Manually measure and record the Pedestal wall temperature at the N/E/S/W positions 
    (consider using iButtons)
    3) Mount the REDS to measuring the pitch angle of the Pedestal (below Azimuth bearing) 
    and aligned with the Azimuth starting position
    4) Start the ACU data recording on the SCU using logging configuration TBD,
    • manually start the REDS data recording (60 seconds @ 1Hz sampling)
    • wait for 60 seconds to reference tilt sensor values
    5) Activate the Azimuth and Elevation axis,
    • wait for 15 seconds to complete the activation process
    6) Command Elevation to 15° at 1°/sec,
    • wait 80 seconds to complete the move
    7) Command Elevation to 85° at 1°/sec,
    • wait 80 seconds to complete the move
    8) Deactivate the axis
    • manually start the REDS data recording (60 seconds @ 1Hz sampling)
    • wait for 60 seconds
    9) Repeat steps 5) and 8) for a total of 5 cycles
    10) Stop the ACU data recording on the SCU
    11) Manually measure and record the Pedestal wall temperature at the N/E/S/W positions (consider using iButtons)
    12) Extract the data on the SCU @ 5Hz sampling
    13) Rename the file to TBD (eg. 6p2p2a_i.csv)
    14) Repeat steps 1) to 13) for Azimuth angles -120,-60,...180
    15) Download the MeerKAT weather station data for the test period
    '''
    for az_angle in azimuths:
        print("\n Azimuth angle: {} \n".format(az_angle))
        #Step 1
        #Manual move dish to starting position
        scu.activate_dmc()
        scu.wait_duration(wait10*3)
        scu.abs_azimuth(az_angle, az_speed)
        scu.wait_duration(wait100)
        el_angle = el_top
        scu.abs_elevation(el_angle, el_speed)
        scu.wait_duration(wait100*8)
        scu.deactivate_dmc()
        scu.wait_duration(wait10*2) 
        
        #Step 2
        #Record pedestal wall temperatures
        print("Record pedestal wall temperatures")

        #Step 3
        #Mount Reds
        print("Mounting REDS")
        #print("ACU Deactivated - time to setup REDS")
        #wait = input("PRESS ENTER TO CONTINUE.")
        print("Script continuing")
        
        #Step 4 - start ACU recording on SCI
        if scu.logger_state() == 'RECORDING':
            print('WARNING, already recording - attempting to stop and start a fresh logger')
            scu.stop_logger()  
            scu.wait_duration(wait10)
        if scu.logger_state() == 'STOPPED':
            print('Starting logger with config: {}'.format(config_name))
            scu.start_logger(config_name)
            scu.wait_duration(wait10)

        i=0
        while i < loop1:
            i+=1
            print("Loop 1: {}".format(i))
            #Step 5
            #Activate az and el
            scu.activate_dmc()
            scu.wait_duration(wait10*3)

            #Move to next az if not there already?
            scu.abs_azimuth(az_angle, az_speed)
            scu.wait_duration(wait100)

            #Step 6
            #az_angle = 200
            el_angle = el_bottom
            scu.abs_elevation(el_angle, el_speed)
            scu.wait_duration(wait100*8)

            #Step 7
            #az_angle = 200
            el_angle = el_top
            scu.abs_elevation(el_angle, el_speed)
            scu.wait_duration(wait100*8)

            #Step 8
            #Deactivate the axis wait 60s record the REDS 
            scu.deactivate_dmc()
            scu.wait_duration(wait10*6) 
            #Do you want to wait for it?
            print("Record REDS")
            #print("ACU Deactivated - time to setup REDS")
            #wait = input("PRESS ENTER TO CONTINUE.")
            print("Script continuing")
            #Step 9 repeat 5 to 8 5 times

        #Step 10 stop recording
        scu.stop_logger()
        scu.wait_duration(wait10)
        print(scu.logger_state())   

        #Step 11 - Manually measure and record the Pedestal wall temperature 
        #Step 12
        print("Extract the data on the SCU @ 5Hz sampling")
        #Step 13    
        print("Rename the file to TBD (eg.6p2p2a_i.csv)")
        scu.save_session('6p2p2a_i_' + str(az_angle), interval_ms = sampling)
    #Step 14 - Repeat steps 1) to 13) for Azimuth angles -120,-60,...180
    #Step 15 - Download the MeerKAT weather station data for the test period
    print("Download the MeerKAT weather station data for the test period")
```
